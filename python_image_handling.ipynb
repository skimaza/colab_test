# 이미지 일부만 잘라서 보기
좌표계에 주의  
수직축이 첫번째 dimension, 수평축이 두번째 dimension
세번째 dimension은 optional (RGB 색상)

```python
img_bw[300:302, 300:302], img_bw[300, 300]
```

```python
type(img_bw)
```

```python
print('img_bw[300:400, 300:400]', '높이 100, 너비 100인 정사각형')
plt.imshow(img_bw[300:400, 300:400], 'gray')
#plt.colorbar()
plt.show()
print('img_bw[350:400, 300:400]', '높이 50, 너비 100인 정사각형')
plt.imshow(img_bw[350:400, 300:400], 'gray')
#plt.colorbar()
plt.show()
print('img_bw[300:400, 300:350]', '높이 100, 너비 50인 정사각형')
plt.imshow(img_bw[300:400, 300:350], 'gray')
#plt.colorbar()
plt.show()
```

죄표축을 보면 왼쪽 상단이 0,0이다.  
즉 plt.imshow로 이미지를 표시하면 np.array의 (0,0)이 왼쪽 위에서 시작하고 첫번째 dimension은 수직으로 내려가며 두번째 dimension은 오른쪽으로 펼쳐진다.
